```python
#import all the required libraries
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
# lets first connect to the webdriver
driver=webdriver.Chrome('chromedriver.exe')
time.sleep(4)
```

```python
url = "https://flight.yatra.com/air-search-ui/dom2/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination=BOM&destinationCountry=IN&flexi=0&flight_depart_date=26%2F10%2F2021&hb=0&noOfSegments=1&origin=DEL&originCountry=IN&type=O&unique=177294764260&version=1.2&viewName=normal"
driver.get(url)
time.sleep(6)
```

```python
airline=[]
Date_of_Journey=[]
Source=[]
Destination=[]
Route=[]
dep_times=[]
arrival_time=[]
duration=[]
stops=[]
layovers=[]
price=[]

time.sleep(3)
#scrapping the required details
start=0
end=4
for page in range(start,end):#for loop for scrapping 4 page
    airlines=driver.find_elements_by_xpath(("//span[@class='i-b text ellipsis']")) #scraping brands name by class name='_2WkVRV'
    for i in airlines:
        airline.append(i.text)#appending the text in Brand list
for page in range(start,end):#for loop for scrapping 4 page
    Sources=driver.find_elements_by_xpath(("//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p"))#scraping brands name by class name='_2WkVRV'
    for i in Sources:
        Source.append(i.text)#appending the text in Brand list        
for page in range(start,end):#for loop for scrapping 4 page
    Destinations=driver.find_elements_by_xpath(("//div[@class='i-b pdd-0 text-left atime col-5']/p[2]"))#scraping brands name by class name='_2WkVRV'
    for i in Destinations:
        Destination.append(i.text)#appending the text in Brand list        
for page in range(start,end):#for loop for scrapping 4 page
    deps_times=driver.find_elements_by_xpath(("//div[@class='i-b pr']"))#scraping brands name by class name='_2WkVRV'
    for i in deps_times:
        dep_times.append(i.text)#appending the text in Brand list        
for page in range(start,end):#for loop for scrapping 4 page
    arrival_times=driver.find_elements_by_xpath(("//div[@class='i-b pdd-0 text-left atime col-5']/p[1]"))#scraping brands name by class name='_2WkVRV'
    for i in arrival_times:
        arrival_time.append(i.text)#appending the text in Brand list        

for page in range(start,end):#for loop for scrapping 4 page
    durations=driver.find_elements_by_xpath(("//div[@class='stop-cont pl-13']/p"))#scraping brands name by class name='_2WkVRV'
    for i in durations:
        duration.append(i.text)#appending the text in Brand list       

for page in range(start,end):#for loop for scrapping 4 page
    stop=driver.find_elements_by_xpath(("//div[@class=' font-lightgrey fs-10 tipsy i-b fs-10']/span"))#scraping brands name by class name='_2WkVRV'
    for i in stop:
        stops.append(i.text)#appending the text in Brand list    

for page in range(start,end):#for loop for scrapping 4 page
    prices=driver.find_elements_by_xpath(("//div[@class='fare-group i-b']"))#scraping brands name by class name='_2WkVRV'
    for i in prices:
        price.append(i.text)#appending the text in Brand list
             
        
df1=pd.DataFrame({'Airlines':airline[:],
                  'Source':Source[:],
                  'Destination':Destination[:],
                  'Dep_times':dep_times[:],
                  'Arrival_time':arrival_time[:],
                  'Duration':duration[:],
                  'Stops':stops[:],
                  'Price':price[:]
                 
                 })
df1
```
