# Details of Sold Car in Bangalore City

```python
# lets first connect to the webdriver
driver=webdriver.Chrome('chromedriver.exe')
time.sleep(4)
```

```python
url = "https://www.cardekho.com/used-cars+in+bangalore"
driver.get(url)
time.sleep(6)
```

```python
brand=[]
model=[]
variant=[]
manuf_year=[]
Driven_KM=[]
fuel=[]
location=[]
No_owners=[]
price=[]

time.sleep(3)
#scrapping the required details
start=0
end=4
for page in range(start,end):#for loop for scrapping 4 page
    brands=driver.find_elements_by_xpath(("//div[@class='gsc_col-xs-7 carsName']/a"))#scraping brands name by class name='_2WkVRV'
    for i in brands:
        brand.append(i.text)#appending the text in Brand list  
    prices=driver.find_elements_by_xpath("//div[@class='price gsc_col-xs-5 text-right']/span")# scraping the discount from the xpath
    for i in prices:
        price.append(i.text)
    models=driver.find_elements_by_xpath(("//div[@class='gsc_col-xs-7 carsName']/div[1]"))#scraping brands name by class name='_2WkVRV'
    for i in models:
        model.append(i.text)#appending the text in Brand list   
    kms=driver.find_elements_by_xpath(("//div[@class='truncate dotlist']/span[1]"))#scraping brands name by class name='_2WkVRV'
    for i in kms:
        Driven_KM.append(i.text)#appending the text in Brand list
    fuels=driver.find_elements_by_xpath(("//div[@class='truncate dotlist']/span[2]"))#scraping brands name by class name='_2WkVRV'
    for i in fuels:
        fuel.append(i.text)#appending the text in Brand list
    variants=driver.find_elements_by_xpath(("//div[@class='truncate dotlist']/span[3]"))#scraping brands name by class name='_2WkVRV'
    for i in variants:
        variant.append(i.text)#appending the text in Brand list
df2=pd.DataFrame({'Brand':brand[:],                 
                 'Price':price[:],
                'Model':model[:],
                'KMS_driven':Driven_KM[:],
                'Fuel':fuel[:],
               'Variant':variant[:]})
df2
```
