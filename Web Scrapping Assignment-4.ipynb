```python
url='https://www.naukri.com/'
driver.get(url)
time.sleep(6)
```

```python
#clicking the explore button
button = driver.find_element_by_xpath("//div[@class='mTxt']")
button.click()
time.sleep(1)
```

```python
search_field_designation=driver.find_element_by_id("qsb-keyword-sugg") #job search bar
search_field_location=driver.find_element_by_id('qsb-location-sugg') #location search bar
search_field_designation.send_keys("Data Science")
```

```python
# clicking the search button
search_button=driver.find_element_by_xpath("//div[@class='search-btn']/button")
search_button.click()
```

```python
# creating empty lists for scraping data
job_title=[]
Designation=[]
job_location=[]
company_name=[]
skill_required=[]

#scraping the job-titles
titles=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
for i in titles:
    if i.text is None :
        job_title.append("--") 
    else:
        job_title.append(i.text)

time.sleep(4)
#scraping the job-location
locations=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
for i in locations:
    if i.text is None :
        job_location.append("--") 
    else:
        job_location.append(i.text)
        
time.sleep(4)        
#scraping the company_name 
companies=driver.find_elements_by_xpath("//div[@class='mt-7 companyInfo subheading lh16']/a")
for i in companies:
    if i.text is None :
        company_name.append("--") 
    else:
        company_name.append(i.text)
company_name[0:2]

time.sleep(4)
#scraping the experience_required 
skill=driver.find_elements_by_xpath("//div[@class='job-description fs12 grey-text']")
for i in skill:
    if i.text is None :
            skill_required.append("--") 
    else:
            skill_required.append(i.text)


time.sleep(4)
# creating the dataframe from the scraped data and taking only first 10 jobs
df=pd.DataFrame({"Skill_required":skill_required[0:10],"company_name":company_name[0:10],"job_location":job_location[0:10],
                "job_title":job_title[0:10]})
df
```

```python
#make empty listes
Name_Year = []  
Genre = [] 
Run_time = [] 
Ratings = []
Votes = []

#scraping the job-titles
name=driver.find_elements_by_xpath("//h3[@class='lister-item-header']")
for i in name:
    if i.text is None :
        Name_Year.append("--") 
    else:
        Name_Year.append(i.text)

time.sleep(4)
#scraping the job-location
genre=driver.find_elements_by_xpath("//span[@class='genre']")
for i in genre:
    if i.text is None :
        Genre.append("--") 
    else:
        Genre.append(i.text)
        
time.sleep(4)        
#scraping the company_name 
run=driver.find_elements_by_xpath("//span[@class='runtime']")
for i in run:
    if i.text is None :
        Run_time.append("--") 
    else:
        Run_time.append(i.text)


time.sleep(4)
#scraping the experience_required 
rating=driver.find_elements_by_xpath("//span[@class='ipl-rating-star__rating']")
for i in rating:
    if i.text is None :
            Ratings.append("--") 
    else:
            Ratings.append(i.text)


time.sleep(4)


voting=driver.find_elements_by_xpath("//span[@class='text-muted']")
for i in voting:
    if i.text is None :
            Votes.append("--") 
    else:
            Votes.append(i.text)
            
time.sleep(4)        

# creating the dataframe from the scraped data and taking only first 10 jobs
df=pd.DataFrame({"Name & Year":Name_Year[0:100],"Genre":Genre[0:100],"Run_time":Run_time[0:100],"Rating":Ratings[0:100],
                "Votes":Votes[0:100]})
df
```
