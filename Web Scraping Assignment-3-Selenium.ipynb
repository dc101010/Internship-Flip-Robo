```python
time.sleep(2)
#locating the search bar
search_bar=driver.find_element_by_class_name("_3704LK")
search_bar.send_keys('moto')
```

```python
time.sleep(2)
#locating the button and clicking it toh search for sunglasses
button=driver.find_element_by_class_name('L0Z3Pu')
button.click()
```

```python
#creating the empty list
brand=[]
price=[]
RAM=[]
Camera=[]
Display=[]
Processor=[]


time.sleep(3)
#scrapping the required details
start=0
end=4
for page in range(start,end):#for loop for scrapping 4 page
    brands=driver.find_elements_by_class_name('_4rR01T')#scraping brands name by class name='_2WkVRV'
    for i in brands:
        brand.append(i.text)#appending the text in Brand list
    prices=driver.find_elements_by_xpath("//div[@class='_30jeq3 _1_WHN1']")# scraping the price from the xpath
    for i in prices:
        price.append(i.text)
    ram=driver.find_elements_by_xpath("//li[@class='rgWa7D']")# scraping the discount from the xpath
    for i in ram:
        RAM.append(i.text)
    camera=driver.find_elements_by_xpath("//li[@class='rgWa7D']")#scraping the list of buttons from the page
    for i in camera:
        Camera.append(i.text)
    display=driver.find_elements_by_xpath("//li[@class='rgWa7D']")#scraping the list of buttons from the page
    for i in display:
        Display.append(i.text)
        
    processor=driver.find_elements_by_xpath("//li[@class='rgWa7D']")#scraping the list of buttons from the page
    for i in processor:
        Processor.append(i.text)
        


df=pd.DataFrame({'Brand':brand[:100],
                'Price':price[:100],
                'Ram':RAM[:100],
                'Camera':Camera[:100],
                'display':Display[:100],
                'Processor':Processor[:100]})
#printing dataframe
df

```

# Question-7

```python
# assign url in the webdriver object
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)
```

```python
url = "https://www.digit.in/top-products/best-gaming-laptops-40.html"
driver.get(url)
time.sleep(6)
```

```python
#creating the empty list
brand=[]
price=[]
OS=[]
Display=[]
Processor=[]
Memory=[]


time.sleep(3)
#scrapping the required details
start=0
end=4
for page in range(start,end):#for loop for scrapping 4 page
    brands=driver.find_elements_by_class_name('Top10-Seller')#scraping brands name by class name='_2WkVRV'
    for i in brands:
        brand.append(i.text)#appending the text in Brand list
    prices=driver.find_elements_by_xpath("//td[@class='smprice']")# scraping the price from the xpath
    for i in prices:
        price.append(i.text)
    os=driver.find_elements_by_xpath("//div[@class='value']")# scraping the discount from the xpath
    for i in os:
        OS.append(i.text)
    display=driver.find_elements_by_xpath("//li[@class='value']")#scraping the list of buttons from the page
    for i in display:
        Display.append(i.text)
    processor=driver.find_elements_by_xpath("//li[@class='value']")#scraping the list of buttons from the page
    for i in processor:
        Processor.append(i.text)
    memo=driver.find_elements_by_xpath("//li[@class='value']")#scraping the list of buttons from the page
    for i in memo:
        Memory.append(i.text)
        


df=pd.DataFrame({'Brand':brand[:7],
                'Price':price[:7],
                'OS':OS[:7],
                'display':Display[:7],
                'Processor':Processor[:7],
                'Memory':Memory[:7]})
#printing dataframe
df

```
